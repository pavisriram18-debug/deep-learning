```python
# Perceptron implementation for AND, OR and NAND Gates

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    # Step Activation Function
    def activation(self, x):
        if x >= 0:
            return 1
        else:
            return 0

    # Prediction Function
    def predict(self, inputs):
        summation = 0
        for i in range(len(inputs)):
            summation += inputs[i] * self.weights[i]
        summation += self.bias
        return self.activation(summation)


# Input combinations
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Bias is fixed
bias = -0.5

# Suitable weights
gates = {
    "AND": [0.5, 0.5],
    "OR": [1, 1],
    "NAND": [-1, -1]
}

# Expected outputs
expected = {
    "AND": [0, 0, 0, 1],
    "OR": [0, 1, 1, 1],
    "NAND": [1, 1, 1, 0]
}

# Verify each gate
for gate in gates:
    print("\n", "=" * 30)
    print(gate, "Gate")
    print("=" * 30)

    perceptron = Perceptron(gates[gate], bias)

    print("X1\tX2\tOutput\tExpected")

    for i, x in enumerate(inputs):
        output = perceptron.predict(x)
        print(f"{x[0]}\t{x[1]}\t{output}\t{expected[gate][i]}")
```
