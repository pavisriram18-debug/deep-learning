```python
import numpy as np

# XOR Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Initial Weights
weights = np.array([1, 1])
bias = -0.5

# Step Activation Function
def step(net):
    if net >= 0:
        return 1
    else:
        return 0

print("XOR Gate using Single Perceptron")
print("-------------------------------------------")
print("X1 X2\tNet Input\tOutput")

for x in X:
    net = np.dot(x, weights) + bias
    output = step(net)
    print(f"{x[0]}  {x[1]}\t{net:.1f}\t\t{output}")
```
