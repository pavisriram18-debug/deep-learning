```python
# Perceptron implementation for OR Gate

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    # Activation function (Step Function)
    def activation(self, x):
        if x >= 0:
            return 1
        else:
            return 0

    # Prediction function
    def predict(self, inputs):
        summation = 0
        for i in range(len(inputs)):
            summation += inputs[i] * self.weights[i]
        summation += self.bias
        return self.activation(summation)


# Suitable weights and bias for OR gate
weights = [1, 1]
bias = -0.5

# Create perceptron object
perceptron = Perceptron(weights, bias)

# OR gate inputs
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Display outputs
print("X1\tX2\tOutput")
for x in inputs:
    output = perceptron.predict(x)
    print(f"{x[0]}\t{x[1]}\t{output}")
```
