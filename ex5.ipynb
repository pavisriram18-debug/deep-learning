```python
import numpy as np

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU Activation Function
def relu(x):
    return max(0, x)

# Perceptron Function
def perceptron(x1, x2, w1, w2, bias, activation):
    net = x1*w1 + x2*w2 + bias
    out = activation(net)
    return net, out

# Input combinations
inputs = [
    (0,0),
    (0,1),
    (1,0),
    (1,1)
]

# AND Gate Weights
w1 = 1
w2 = 1
bias = -1.5

# ---------------- Sigmoid ----------------
print("===== Sigmoid Activation =====")
print("X1 X2\tNet\tSigmoid Output")

for x1, x2 in inputs:
    net, out = perceptron(x1, x2, w1, w2, bias, sigmoid)
    print(f"{x1}  {x2}\t{net:.1f}\t{out:.4f}")

# ---------------- ReLU ----------------
print("\n===== ReLU Activation =====")
print("X1 X2\tNet\tReLU Output")

for x1, x2 in inputs:
    net, out = perceptron(x1, x2, w1, w2, bias, relu)
    print(f"{x1}  {x2}\t{net:.1f}\t{out:.1f}")
```
