```python
# Perceptron implementation for NOR Gate

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    # Activation function (Step Function)
    def activation(self, x):
        if x >= 0:
            return 1
        else:
            return 0

    # Prediction function
    def predict(self, inputs):
        summation = 0
        for i in range(len(inputs)):
            summation += inputs[i] * self.weights[i]
        summation += self.bias
        return self.activation(summation)


# Suitable weights and bias for NOR gate
weights = [-1, -1]
bias = 0.5

# Create perceptron object
perceptron = Perceptron(weights, bias)

# NOR gate inputs
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Display outputs
print("X1\tX2\tOutput")
for x in inputs:
    output = perceptron.predict(x)
    print(f"{x[0]}\t{x[1]}\t{output}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Input points
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([1,0,0,0])

# Plot points
for i in range(len(X)):
    if Y[i] == 1:
        plt.scatter(X[i][0], X[i][1], color='green', s=100, label='Output = 1' if i==0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='red', s=100, label='Output = 0' if i==1 else "")

# Decision boundary: x1 + x2 = 0.5
x = np.linspace(-0.2, 1.2, 100)
y = 0.5 - x

plt.plot(x, y, 'b-', label='Decision Boundary')

plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Decision Boundary for NOR Gate")
plt.grid(True)
plt.legend()
plt.show()
```
