```python
import numpy as np

# Training data for AND Gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target output
T = np.array([0, 0, 0, 1])

# Initial Parameters
learning_rate = 0.2
weights = np.array([0.3, 0.1])
bias = 0

# Step Activation Function
def step(net):
    if net >= 0:
        return 1
    else:
        return 0

print("Perceptron Training for One Epoch")
print("-" * 70)
print("Input\tTarget\tNet\tOutput\tError\tUpdated Weights\tUpdated Bias")

# One Epoch
for i in range(len(X)):

    net = np.dot(X[i], weights) + bias

    output = step(net)

    error = T[i] - output

    # Weight Update Rule
    weights = weights + learning_rate * error * X[i]

    # Bias Update Rule
    bias = bias + learning_rate * error

    print(f"{X[i]}\t {T[i]}\t {net:.2f}\t {output}\t {error}\t {weights}\t {bias:.2f}")

print("\nFinal Weights :", weights)
print("Final Bias :", bias)
```
